```python
# RANDOM FOREST 
# RANDOM FOREST IS A COMBINATION OF DECISION TREES WHERE IN THE DECISION IS TAKEN BY VARIOUS DECISION TREES FOR THE
# SAME DATASET AND THE DECISION CHOSEN BY THE MAJORITY IS THE PREDICTED VALUE
# THE DATASETS ARE RANDOMLY SPLIT AND EACH OF THE DECISION TREE IS TRAINED DIFFERENT RANDOMLY SPLIT PORTIONS OF THE
# THE DATASETS HENCE THE WORD 'RANDOM'
# SINCE A LOT OF DECISION TREES ARE TRAINED AND USED FOR PREDICTION HENCE THE WORD 'FOREST'
```

```python
import pandas as pd
import numpy as np 
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
# THE TARGET LABELS ARE 0-9
digits.target_names
```

```python
# THE TARGET IS 0-9 TILL FOR 1798 ROWS
digits.target
```

```python
df=pd.DataFrame(digits.data)
df
```

```python
# LETS APPEND THE TARGET COLUMN TO THE DATAFRAME 
df['Target']=pd.DataFrame(digits.target)
df
```

```python
# NOW LETS SPLIT THE DATASET INTO TRAINING AND TESTING SET USING TRAIN_TEST_SPLIT
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df.drop('Target',axis='columns'),df.Target,test_size=0.2)
```

```python
X_train
```

```python
len(X_train)
```

```python
# THE RANDOM FOREST CLASSIFIER IS IN THE ENSEMBLE MODULE OF SKLEARN
# ENSEMBLE CONSISTS OF MACHINE LEARNING MODEL WITH MULTIPLE MODELS/ALGORITHMS 
# SINCE RANDOM FOREST IS JUST A COMBINATION OF DECISION TREES IT IS STORED IN ENSEMBLE OF SKLEARN
from sklearn.ensemble import RandomForestClassifier
# LETS CREATE AN OBJECT OF THE RANDOMFORESTCLASSIFIER
# THE CRITERION CAN EITHER BE 'gini' OR 'entropy' or 'log_loss' AND DECIDES THE WAY IN WHICH THE DECISION TREE IS 
# PLOTTED
# THE n_estimators CAN BE OF ANY NUMBER AND IT DECIDES THE NUMBER OF DECISION TREE IN THE RANDOM FOREST
model=RandomForestClassifier()
model.fit(X_train,Y_train)
```

```python
# NOW LETS CHECK THE SCORE/PERFORMANCE OF THE MODEL AND IT HAS SCORED AN ACCURACY OF 96% 
model.score(X_test,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier 
model=RandomForestClassifier(criterion='entropy')
model.fit(X_train,Y_train)
```

```python
# SO WITH CRITERION AS 'entropy IT PERFORMANCE INCREASE TO 97 %
model.score(X_test,Y_test)
```

```python
# SO IT THE PERFORMANCE ROSE FROM 97.22% TO 97.77% WHEN WE CHANGED THE n_estimators from 100 TO 1000 AND CRITERION
# TO 'entropy'
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=1000,criterion='entropy')
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
