# IRIS FLOWER CLASSIFICATION USING LOGISTIC REGRESSION 



Using sklearn.datasets iris flower dataset to train your model using logistic regression. You need to figure out accuracy of your model and use that to predict different samples in your test dataset. In iris dataset there are 150 samples containing following features,

Sepal Length
Sepal Width
Petal Length
Petal Width
Using above 4 features you will clasify a flower in one of the three categories,

Setosa
Versicolour
Virginica

```python
import pandas as pd
import numpy as np 
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
```

```python
dataset=load_iris()
```

```python
dir(dataset)
```

```python
dataset
```

```python
# LETS PLOT THE DATASET USING MATSHOW FUNCTION OF MATPLOTLIB
plt.gray()
for i in range(5):
    plt.matshow([dataset.data[i]])
```

```python
# NOW LETS SPLIT THE TRAINING AND TESTING SET FROM THE DATASET
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(dataset.data,dataset.target, test_size=0.2)
```

```python
# LETS TRAIN THE MODEL ON THE SPLIT DATASET
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
# THE MODEL ACHEIVED AN ACCURACY OF 93 %
model.score(X_test,Y_test)
```

```python
# LETS SEE THE AREAS WHERE THE MODEL FAILED TO PREDICT IT RIGHT WE ARE USING CONFUSION MATRIX 
from sklearn.metrics import confusion_matrix
Y_predicted=model.predict(X_test)
cm=confusion_matrix(Y_predicted,Y_test)
cm
```

```python
# LETS PLOT THE CONFUSION MATRIX USING SEABORN 
import seaborn as sn 
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
# SO THE ONLY MISTAKE THE MODEL MADE WAS PREDICTING THE VALUE 2 AS 1 FOR 2 TIMES THE REST ALL ARE RIGHTLY PREDICTED
```
