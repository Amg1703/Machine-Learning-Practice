# K FOLD CROSS VALIDATION

```python
# SO WE TRAIN IN THE MODEL IN 3 WAYS :- 
# 1) WHEREIN YOU TRAIN AND TEST THE MODEL WITH COMPLETE 100% OF THE DATASET BUT THIS CAN CAUSE OVERFITTING AS THE 
# MODEL WILL NOT LEARN AND IT IS NOT FAIR TO TEST THE MODEL ON THE SAME DATASET IT WAS TRAINED SINCE IT KNOWS THE 
# ANSWER
# 2) HERE WE SPLIT THE DATASET INTO TRAINING AND TESTING BUT THE DATASET MIGHT BE SPLIT IN SUCH A WAY THAT ONE 
# THE DATASET OF CATEGORY OF TARGET VARIABLE WAS TRAINED MORE AND THE OTHER LESS THIS CAN CAUSE FLUCTUATIONS IN THE
# PERFORMANCE OR ACCURACY OF THE MODEL AS EACH TIME IT IS TRAINED AND TESTED ON DIFFERENT DATASET
# 3) K FOLD - THIS IS WHERE WE TAKE THE DATASET AND SPLIT INTO FOLD AND A SET OF FOLDS ARE USED FOR TRAINING AND 
# REST FOR TESTING AND THIS IS DONE IN LOOP/ITERATION UNTIL ALL THE FOLDS ARE USED FOR TRAINING AND TESTING 
# THEN IN THE END THE AVERAGE SCORE/PERFORMANCE OF THE MODEL IS GIVEN AS THE PERFORMANCE OF THE MODEL
# THIS IS MORE FAIR SINCE THE MODEL IS TRAINED AND TESTED ACCROSS ALL THE DATASET AND ONLY THEN EVALUATED
# FOR EXAMPLE IF WE HAVE DATASET WITH 100 VALUES AND WE SPLIT IT INTO 5 FOLDS THEN :-
# 1ST ITERATION - THE 1ST FOLD IS RESERVED FOR TESTING AND 2-5 ARE USED FOR TRAINING 
# 2ND ITERATION - THE 2ND FOLD IS RESERVED FOR TESTING AND 1,3-5 ARE USED FOR TRAINING 
# 3RD ITERATION - THE 3RD FOLD IS RESERVED FOR TESTING AND 1-2,4-5 ARE USED FOR TRAINING 
# 4TH ITERATION - THE 4TH FOLD IS RESERVED FOR TESTING AND 1-3,5 ARE USED FOR TRAINING 
# 5TH ITERATION - THE 5TH FOLD IS RESERVED FOR TESTING AND 1-4 ARE USED FOR TRAINING 
```

```python
import pandas as pd
import numpy as np 
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
#digits.feature_names
```

```python
df=pd.DataFrame(digits.data)
df
```

```python
df['Target']=pd.DataFrame(digits.target)
df
```

```python
# LETS SPLIT THE DATASET INTO TRAINING AND TESTING TO TEST HOW THE MODEL'S PERFORMANCE FLUCTUATES WITH EACH TIME
# WE RUN THIS CELL AGAIN CAUSING THE ALTERING TRAINING AND TESTING DATASET
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.drop('Target',axis='columns'),df.Target,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
# NOW LETS USE KFOLD FROM THE MODEL SELECTION MODULE OF SKLEARN
# n_splits PARAMETER DETERMINES THE NUMBER OF FOLDS KFOLD FUNCTION HAS TO PERFORM ON THE DATASET
from sklearn.model_selection import KFold
fold=KFold(n_splits=3)
# SO WHAT THE KFOLD FUNCTION DOES IS IT GOES THROUGH THE ARRAY OR THE DATASET AND INDEXES THE DATASET OR ARRAY
# AND UNIFORMLY JUMBLES INDEX WHICH WE CAN ASSIGN TO THE TRAINING AND TESTING SET SUCH THAT IT COVERS THE
# COMPLETE DATASET
# 
for train_index, test_index in fold.split([1,2,3,4,5,6,7,8,9]):
    print(train_index, test_index)
```

```python
# LETS DEFINE A FUNCTION TO PERFORM TRAINING AND CHECKING THE SCORE OF THE MODEL 
# THIS FUNCTION WILL TAKE IN THE PARAMETERS MODEL, X_TRAIN, X_test,Y_train,Y_test AND TRAIN THE MODEL USING 
# X_TRAIN AND Y_TRAIN AND WILL RETURN THE SCORE OF THE MODEL USING X_TEST AND Y_TEST
def score_model(model,X_train,X_test,Y_train,Y_test):
    model.fit(X_train,Y_train)
    return model.score(X_test,Y_test)
print(score_model(LogisticRegression(),X_train,X_test,Y_train,Y_test))
print(score_model(SVC(),X_train,X_test,Y_train,Y_test))
print(score_model(RandomForestClassifier(),X_train,X_test,Y_train,Y_test))
```

```python
# STRATIFIEDKFOLD - DOES A BETTER DISTRIBUTION OF FOLDS BY SPLITTING THEM UNIFORMLY AND IS VERY WELL SUITED FOR 
# DATASET WITH IMBALANCED DATA DISTRIBUTION
# SO IN OTHERWORDS STRATIFIEDKFOLD IS A BETTER VERSION OF KFOLD 
from sklearn.model_selection import StratifiedKFold
fold=StratifiedKFold(n_splits=3)
LogisticRegression_score=[]
SVM_score=[]
RandomForestClassifier_score=[]
# SO WE 
for training_index,testing_index in fold.split(digits.data,digits.target): 
    X_train, X_test, Y_train, Y_test = digits.data[training_index], digits.data[testing_index], \
                                        digits.target[train_index], digits.target[testing_index]
    # SO THE SCORE MODEL RETURNS THE SCORE OF THE MODEL WHICH IS THEN APPENDED TO EMPTY ARRAY DEFINED ABOVE
    LogisticRegression_score.append(score_model(LogisticRegression(),X_train,X_test,Y_train,Y_test))
    SVM_score.append(score_model(SVC(),X_train,X_test,Y_train,Y_test))
    RandomForestClassifier_score.append(score_model(RandomForestClassifier(),X_train,X_test,Y_train,Y_test))
print(LogisticRegression_score)
```

```python
# cross_validation_score() - THE ABOVE OPERATION OF PERFORMING KFOLD AND STARTIFIED KFOLD OPERATION ON THE DATASET ALONG WITH THE SCORE_MODEL 
# FUNCTION CAN BE PERFORMED BY THE CROSS_VALIDATION_SCORE FUNCTION
from sklearn.model_selection import cross_val_score
# JUST LIKE SCORE_MODEL WE DEFINED ABOVE ALL WE HAVE TO PASS IS THE MODEL(), digits.data, digits.target AS PARAMETER
print(cross_val_score(LogisticRegression(),digits.data,digits.target))
print(cross_val_score(SVC(),digits.data,digits.target))
print(cross_val_score(RandomForestClassifier(),digits.data,digits.target))
```

```python
# WE CAN NOT ONLY USE THE CROSS VALIDATION SCORE FOR CHECKING PERFORMANCE OF DIFFERENT MODEL BUT ALSO 
# HOW EACH PARAMETER CHANGES THE PERFORMANCE OF THE MODEL 
# THIS IS CALLED PARAMETER FINE TUNING 
print(cross_val_score(RandomForestClassifier(n_estimators=150,criterion='entropy'),digits.data,digits.target))
```

```python
print(cross_val_score(RandomForestClassifier(n_estimators=50,criterion='gini'),digits.data,digits.target))
```
