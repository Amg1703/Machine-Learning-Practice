```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('homeprices (2).csv')
df
print(type(df))
```

```python
FOR EXAMPLE IN THE ABOVE DATASET THERE ARE 3 TOWNS MONROE TOWNSHIP, WEST WINDSOR AND ROBINSVILLE EACH WITH ITSÂ¶
OWN SET OF PRICE ACCORDING AREA
NOW SINCE COMPUTER DOESNT UNDERSTAND TEXTUAL DATA WE CAN USE NUMERICAL DATA LIKE 1,2,3
BUT NUMBERING THE TOWNS AS 1,2,3 CAN CAUSE CONFUSION FOR THE MACHINE LEARNING MODEL SINCE IT WILL ASSUME
ROBINSVILLE > WEST WINDSOR > MONROE AS IT IS 3>2>1
THESE COME UNDER CATEGORICAL VALUES WHEREIN THERE ARE 2 TYPES NAMELY
1) NOMINAL VALUE - WHERE ALL THE GROUPS ARE ON AN EQUAL LEVEL AND ONE GROUP IS NOT SUPERIOR OVER OTHER LIKE COLOURS
RED , GREEN, VALUE OR TOWNSHIPS - BESANT NAGAR, ADYAR, GUINDY
2) ORDINAL VALUE - WHERE ALL THE GROUPS ARE IN COMPETITIVE ORDER LIKE AND GROUP IS SUPERIOR OVER OTHER LIKE HIGH,
MEDIUM, LOW ... GOOD,BAD AND UGLY OR SATISFIED, NEUTRAL AND DISATISFIED
```

```python
dummies=pd.get_dummies(df.town,dtype=int)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
merged.columns
```

```python
# DUMMY VARIABLE TRAP 
# WHENEVER ONE VARIABLE CAN BE REPRESENTED BY ANOTHER IT CALLED MULTI-COLINEAR AND WHENEVER YOU HAVE MULTI-COLINEAR
# VARIABLE IT CREATES THE DUMMY VARIABLE TRAP 
# SO WE CAN DELETE ONE COLUMN OF THE DUMMY VARIABLE AS 2 COLUMNS IS ENOUGH THE OTHER 1 COLUMN CAN BE REPRESENTED AS 
# THE OPPOSITE/NEGATIVE OF THE 2 
# FOR EXAMPLE WHEN BOTH THE COLUMNS MONROE TOWNSHIP AND ROBINSVILLE HAS 0 AS THE VALUE FOR A ROW THEN IT MEANS IT IS
# WEST WINDSOR
# THE SKLEARN LINEAR REGRESSION WILL AUTOMATICALLY DROP THE MULTI-COLINEAR COLUMN VARIABLE BUT IT IS GOOD PRACTICE
# TO DROP THE COLUMN MANUALLY
```

```python
final_df=merged.drop(['town','west windsor'], axis='columns')
final_df
```

```python
# LETS CREATE A SEPERATE OBJECT FOR THE INPUT DATAFRAME
X = final_df.drop(['price'], axis='columns')
X
```

```python
# LETS CREATE A SEPERATE OBJECT FOR PRICE 
Y = final_df.price
Y
```

```python
# WE ARE TRAINING THE LINEAR REGRESSION MODEL ON THE STORED OBJECTS
model=LinearRegression()
model.fit(X,Y)
```

```python
# ASKING TO PREDICT 3550 SQT AREA PRICE IN MONROE TOWNSHIP
model.predict([[3550,1,0]])
```

```python
# ASKING TO PREDICT 2000 SQT AREA PRICE WEST WINDSOR 
model.predict([[2000,0,0]])
```

```python
# WE CAN DO THE SAME USING PREPROCESSING FUNCTION IN SKLEARN LIKE LABEL ENCODER AND ONEHOTENCODER
# WHERE THE LABEL ENCODER IS USED TO CREATE THE LABELS FOR THE DATAFRAME BASED ON SPECIFIC MENTIONED COLUMNS IN THE
# DATAFRAME
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
# USE fit_transform FUNCTION IN OF LabelEncoder TO TRANSFORM THE COLUMN INTO CATEGORICAL LABELS
df['town']=LE.fit_transform(df.town)
df
```

```python
print(type(df))
```

```python
# NOW LETS CONVERT THE DATAFRAME INTO AN ARRAY 
X=df[['town','area']]
X
```

```python
Y = df.price
Y
```

```python
# OneHotEncoder()
# NOW TO ADD THE DUMMY VARIABLE COLUMN FROM THE LABELED COLUMNS WE USE ONEHOTENCODER
# SINCE WE DONT WANT THE ENTIRE X DATAFRAME TO BE CATEGORICALLY LABELLED AND WE ONLY WANT SPECIFIC COLUMNS
# WE USE COLUMN TRANSFORMER AND ONEHOTENCODER
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), ['town'])  # 'town' is the name of the column
    ],
    remainder='passthrough'  # Leave 'area' as-is
)

X=ct.fit_transform(X)
X
```

```python
# TO AVOID DUMMY VARIABLE LETS DROP THE FIRST COLUMN 
X=X[:,1:] # THIS WILL SELECT ALL ROWS : AND 1: WILL CHOOSE COLUMNS AFTER 1ST COLUMNS
X
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,Y)
```

```python
# PREDICTS THE VALUE OF 3500 SQFT OF AREA OF MONROW TOWNSHIP AS IT IS 0,0 
model.predict([[0,0,3500]])
```
