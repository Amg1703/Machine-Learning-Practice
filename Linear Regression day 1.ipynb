```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plot
```

```python
# LINEAR REGRESSION WORKS BY UTITLLISING LINEAR EQUATION ON THE DATASET WHEREIN THE LINEAR RELATIONSHIP BETWEEN THE
# DEPENDENT VAIRABLES WHICH ARE THE METRICS AND THE INDEPENDENT VARIABLE WHICH IS THE VALUE TO BE PREDICTED
# LINEAR EQUATION Y = MX + C 
# SO LINEAR EQUATION MAPS THE RELATIONSHIP BETWEEN Y (INDEPENDENT VARIABLE) AND X (DEPENDENT VARIABLE/ METRICS) 
# BY FINDING THE VALUE OF M AND C 
```

```python
df=pd.read_csv('homeprices.csv')
df
```

```python
# FIRST LETS PLOT THE GRAPH OF THE VALUES USING MATPLOTLIB
%matplotlib inline
plot.scatter(df.area,df.price)
```

```python
# SO LINEAR REGRESSION DRAWS VARIOUS COMBINATION OF LINES BETWEEN EACH OF THE POINTS AND FIND THE ERROR WHICH IS 
# THE DISTANCE BETWEEN THE LINE AND THE MISSING POINTS THROUGH WHICH THE LINE HAS NOT PASSED THROUGH 
# SO THE LINE WITH THIS MINIMAL VALUE OF THIS ERROR IS CHOSEN AS THE LINE 
# AND THE SLOPE WHICH IS THE M VALUE AND C VALUE OF THIS LINE IS TAKEN TO FORM THE LINEAR EQUATION FOR THE DATASET
```

```python
# NOW LETS MAKE THE GRAPH MORE APPEALING BY ALTERING THE POINTS
plot.xlabel('Area (in Sqft)')
plot.ylabel('Prices (in US $)')
plot.scatter(df.area,df.price, color='red', marker='+')
```

```python
# NOW LETS CREATE AN OBJECT FOR LINEAR REGRESSION 
reg=linear_model.LinearRegression()
# fit(DataFrame) IN ORDER TO TRAIN THE MODEL USING OUR DATAFRAME
reg.fit(df[['area']],df.price)
```

```python
# predict() - TO GIVE THE PREDICTED VALUE BY INPUTTING THE VALUE IN THE MODEL 
reg.predict([[5000]])
```

```python
# coef - TO GIVE THE COEFFICIENT VALUE OF THE LINEAR EQUATION COMPREHENDED BY THE MODEL WHICH IS Y=MX+C WHICH IS THE
# M VALUE
reg.coef_
```

```python
# intercept - TO GET THE INTERCEPT VALUE OF THE LINEAR EQUATION COMPREHENDED BY THE MODEL WHICH IS Y=MX+C WHICH IS THE
# C VALUE
reg.intercept_
```
