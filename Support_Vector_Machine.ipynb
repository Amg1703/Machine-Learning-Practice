```python
# SUPPORT VECTOR MACHINE IS USED FOR CLASSIFICATION PROBLEM WHEREIN DRAWS A LINE OR PLANE BETWEEN THE TWO OR MORE
# FEATURES OR DIMENSIONS
# SO THIS LINE OR PLANE DRAWN IS TO BE IN SUCH A WAY THAT DATASETS PLOTTED HAVE MAXIMUM DISTANCE FROM THE LINE
# THE DATASETS PLOTTED ARE CALLED THE SUPPORT VECTOR MACHINES AND ONLY THE NEAREST ARE CHOSEN FOR PLOTTING THE LINE
# THE LINE OR PLANE CHOSEN SHOULD HAVE MAXIMUM DISTANCE WITH THE VECTOR MACHINES BECAUSE THAT SHOWS MODEL 
# CLASSIFIES WELL DISTINCTIVELY BETWEEN THE 2 DATAS
# WHEN THE THERE ARE N DIMENSION/FEATURES THEN WE USE HYPERPLANE WHEREIN THE MODEL DRAWS THE HYPERPLANE BETWEEN
# SUPPORT VECTOR MACHINE WITH MAXIMUM DISTANCE FROM THE SUPPORT VECTOR MACHINES 
```

```python
# GAMMA :-
# GAMMA IS TO DO WITH THE NUMBER OF SUPPORT VECTOR MACHINES INVOLVED NORMALLY WE ONLY CHOOSE THE CLOSE AND NEIGHBOURING
# SUPPORT VECTOR MACHINES BUT WITH WHICH WE CAN DRAW A CURVER OR A MORE FITTING LINE WITH MAXIMUM DISTANCE FROM
# NEIGHBOURING SUPPORT VECTOR MACHINES THIS IS CALLED HIGH GAMMA
# BUT IF WE CONSIDER MORE AND MORE SUPPORT VECTOR MACHINES THEN WE MIGHT HAVE MORE STRAIGHT LINE WITH LESS DISTANCE
# FROM THE SUPPORT VECTOR MACHINES THAT IS LESS FITTING THIS IS CALLED LOW GAMMA 
# HIGH GAMMA :- 
# MORE ACCURACY AND PRECISE WITH CURVED LINE AND MORE FITTING 
# BUT MORE COMPUTATIONAL RESOURCES 
# LOW GAMMA :- 
# COMPUTATIONAL MORE EFFICIENT SINCE IT IS A STRAIGHT LINE
# LESS ACCURACY 
# BOTH APPROACHES MIGHT BE RIGHT AND UPTO THE INDIVIDUAL AS WITH LOW GAMMA IT IS COMPUTATIONALY LESS EFFICIENT AND 
# WITH HIGH GAMMA IT IS MORE PRECISE 
```

```python
# REGULARIZATION :-
# WHEN YOU DRAW THE LINE OR PLANE ALL ACCORDING TO THE NEIGHBOURING SUPPORT VECTOR MACHINES THAT THE SOLE AIM IS 
# ONLY TO CLASSIFY BETWEEN THE 2 SIDES THAT YOUR LINES BECOMES TOO CURVED AND OVERFITTED FOR THE DATASET THIS IS CALLED
# HIGH REGULARIZATION
# LOW REGULARIZATION IS WHEN THE LINE OR PLANE DRAWN IS STRAIGHT EVEN THOUGH SOME VECTOR MACHINES MIGHT BE TOO CLOSE
# OR EVEN ON THE OPPOSITE SIDE OF THE THE LINE OR PLANE THIS CAN CAUSE THE MODEL TO PRODUCE SOME ERRORS BUT ITS OKAY
```

```python
# KERNELS 
# KERNEL IS TRANSFORMATION BY ADDING ANOTHER DIMENSION 
# IN SOME CASES THE DATA MIGHT BE TOO MIXED UP AND MIGHT BE HARD TO CLASSIFY JUST BY A SINGLE LINE OR PLANE
# IN SUCH CASE YOU CREATE AN EXTRA DIMENSION SO THAT YOU CAN DIFFERENTIATE THE DATA DISTINCTIVELY ON THIS DIMENSION
```

```python
import pandas as pd
import numpy as np 
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
dir(dataset)
```

```python
dataset
```

```python
# SO FROM THE DATASET WE CAN SEE THAT DATA, FEATURE NAMES AND TARGET ARE THE ONLY COLUMNS THAT ARE NEEDED
# NOW LETS CONVERT THE GIVEN DATASET INTO A DATAFRAME
df=pd.DataFrame(dataset.data, columns = dataset.feature_names)
df
```

```python
# NOW LETS APPEND THE TARGET COLUMN TO THE DATAFRAME
df['target']=pd.DataFrame(dataset.target)
df
```

```python
# NOW FOR REFERENCE LETS ALSO ADD THE FLOWER NAME COLUMN ALONG WITH THE TARGET COLUMN
# NOW WE ARE CREATING A NEW COLUMN FROM A REFERENCE COLUMN THIS CAN BE DONE BY APPLY FUNCTION 
# THE LAMDA IS JUST MINI VERSION OF AN ACTUAL FUNCTION 
# SO THE CODE ITERATES THROUGH THE TARGET COLUMN OF DATAFRAME(df.target)
# AND THE TARGET VALUE IS STORED IN X ACCORDING TO WHICH THE TARGET NAME CONSISTING OF 'setosa', 'versicolor', 
# 'virginica' IS INDEXED FOR EXAMPLE IF THE 3 ROW IS INDEXED WITH TARGET VALUE 0 THEN THE TARGET NAME WITH INDEX 0
# IS INDEXED AND THEN STORED IN THE FLOWER NAME COLUMN
df['Flower_Name']=df.target.apply(lambda x: dataset.target_names[x])
df
```

```python
# NOW LETS CREATE SUB DATAFRAME
# df.groupby('Flower_Name')
```

```python
# NOW LETS DIVIDE THE DATAFRAME INTO SUB DATAFRAMES FOR PLOTTING EACH FLOWER AND ITS FEATURED SEPERATELY
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
# LETS PLOT THE SUBDATAFRAME USING MATPLOTLIB USING SEPAL LENGHT AND SEPAL WIDTH
# SO WE CAN SEE THE DATAFRAME 1 AND 2 CAN BE CLEARLY SPERATED WHILE THE DATAFRAME 3 IS QUITE MIXED
from matplotlib import pyplot as plt
%matplotlib inline
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'], marker='+', color='red')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], marker='.', color='green')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'], marker='*', color='blue')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
# NOW LETS PLOT THE petal length (cm) AND petal width (cm) OF THE FLOWER
# SO WE CAN SEE THAT BASED ON PETAL LENGTH AND WIDTH THESE 3 FLOWERS CAN BE SEPERATED PERFECTLY SO SUPPORT VECTOR 
# MACHINE WOULD BE THE RIGHT CHOICE FOR THIS PROBLEM
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], color='red', marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], color='green', marker='.')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'], marker='*', color='blue')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```

```python
X=df.drop(['target','Flower_Name'], axis='columns')
X
```

```python
Y=df.drop(['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)','Flower_Name'], axis='columns')
Y
```

```python
# NOW LETS SPLIT THE DATASET INTO TRAINING AND TESTING 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

```python
X_train
```

```python
X_test
```

```python
# NOW LETS CREATE AN OBJECT OF THE MODEL
# SO THE C IS REGULARIZATION WHOSE DEFAULT VALUE IS 3 BUT CAN BE SET TO ANY VALUE
# KERNEL'S DEFAULT VALUE IS 'rbf' BUT CAN BE SET TO {'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'}
# GAMMA'S DEAULT VALUE IS  'scale' BUT CAN BE SET 'auto' or 'scale'
from sklearn.svm import SVC
model=SVC()
model.fit(X_train,Y_train)
```

```python
# WE HAVE ACHEIVED AN ACCURACY OF 94% 
model.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC
model=SVC(C=100)
model.fit(X_train,Y_train)
```

```python
# SO INCRESING REGULARIZATION TO 100 THE ACCURACY DROPPER
model.score(X_test,Y_test)
```

```python
from sklearn.svm import SVC
model=SVC(gamma='auto')
model.fit(X_train,Y_train)
```

```python
# CHANING THE GAMMA VALUE TO AUTO ALSO GIVES THE SAME VALUE
model.score(X_test,Y_test)
```

```python
# CHANGING THE KERNEL TO LINEAR POLY AND RBF GAVE THE SAME ACCURACY OF 94 % BUT SIGMOID DROPPED TO 26%
# IT DIDNT SUPPORT PRECOMPUTED AS PRECOMPUTED REQUIRED SQUARE SHAPED DATAFRAMES
from sklearn.svm import SVC
model=SVC(kernel='linear')
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
