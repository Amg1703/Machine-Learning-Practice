```python
import numpy as np 

def gradient_descent(x,y):
    m_curr=b_curr=0 # M IS THE SLOPE VALUE AND B IS THE INTERCEPT VALUE IS INITIALLIZED TO 0
    iterations=10 # INCREASING THE NUMBER OF ITERATIONS WILL HELP THE GRADIENT FUNCTION 
    # TO REACH CLOSER TO GLOBAL MAXIMA
    # HERE WITH 10000 ITERATIONS WE CAN SEE THAT THE COST VALUE REACHES A CONSTANT VALUE WHICH
    # MEANS THAT GLOBAL MINIMA IS REACHED
    n=len(x) # THE LENGTH OF THE ARRAY OR THE LIST OF ELEMENTS OF X AND Y 
    learning_rate=0.01 # THIS DECIDES THE SIZE OF THE STEP TAKEN FOR EACH ITERATION
    # HIGHER LEARNING RATE CAN HELP REACH THE MINIMA FASTER BUT ALSO THERE IS CHANCE TO SKIP THE
    # MINIMA POINT AND GO UP THE GRAPH ON THE OTHERSIDE BUT CAN REACH GLOBAL MINIMA IN 
    # FEWER ITERATIONS
    # LOWER LEARNING RATE WILL MOVE FINE AND WILL JUMP VALUES SLOWLY AND MOVE SLOWLY TO THE 
    # GLOBAL MINIMA BUT WILL BE MORE PRECISE 
    for i in range(iterations):
        y_predicted=m_curr*x+b_curr
        cost=(1/n)*sum(val**2 for val in (y-y_predicted))# THE Y-Y PREDICTED VALUE IS
        # ASSIGNED TO THE VAL VARIABLE AND THIS VAL VALUE IS SQAURED TO AVOID NEGATIVE VALUES
        # LESSER THE COST MEANS SLOPE AND INTERCEPT ARE CLOSE TO THE GLOBAL MINIMA
        md=-(2/n)*sum(x*(y-y_predicted)) # THIS IS THE DERIVATIVE WHICH IS THE SLOPE VALUE
        bd=-(2/n)*sum(y-y_predicted) # THIS IS THE DERIVATIVE WHICH IS THE INTERCEPT VALUE
        m_curr=m_curr-learning_rate*md # THIS FORMULA DECIDES THE STEP TAKEN FOR EACH STEP
        b_curr=b_curr-learning_rate*bd
        print('m{},b{},cost{},iteration{}'.format(m_curr,b_curr,cost,i))
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

gradient_descent(x,y)
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    iterations=1000
    n=len(x)
    learning_rate=0.01
    for i in range(iterations):
        y_predicted=m_curr*x+b_curr
        cost=(1/n)*sum([val**2 for val in (y-y_predicted)])
        md=(2/n)*sum(x*(y-y_predicted))
        bd=(2/n)*sum(y-y_predicted)
        m_curr=m_curr-learning_rate*md
        b_curr=b_curr-learning_rate*bd
        print('m{},b{},cost{},iteration{}'.format(m_curr,b_curr,cost,i))
gradient_descent(x,y)
```
