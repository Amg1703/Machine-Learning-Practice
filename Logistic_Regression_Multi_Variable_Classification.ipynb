# LOGISITIC REGRESSION MULTIVARIABLE CLASSFICATION - HAND WRITTEN DIGIT RECOGNITION

```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
from sklearn.datasets import load_digits # THE LOAD_DIGITS IS A DATASET CONSISTING OF PREDEFINED DATASET
```

```python
# LET'S CREATE AN OBJECT FOR THE DATASET
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.data[0:5]
```

```python
# THE IMAGES ARE ALSO STORED AS ARRAYS OF NUMBERS
digits.images[0:5]
```

```python
# SINCE THE IMAGE DATASET IS IN THE FORM OF NUMERICAL ARRAY WE ARE USING THE MATSHOW FUNCTION TO PLOT IT AS AN IMAGE
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
# NOW LETS SPLIT THE DATASET INTO TRAINING AND TESTING 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(digits['data'],digits['target'], test_size=0.2, random_state=10)
```

```python
# NOW LETS IMPORT THE LOGISTIC REGRESSION MODEL
from sklearn.linear_model import LogisticRegression
# LETS CREATE AN OBJECT FOR THE MODEL
model=LogisticRegression()
# LETS TRAIN THE MODEL USING THE FIT FUNCTION
model.fit(X_train,Y_train)
```

```python
# NOW LETS MAKE PREDICTION USING THE TRAINED LOGISTIC REGRESSION MODEL 
model.predict(X_test)
```

```python
plt.matshow(digits.images[67])
# SO THE 67 TH INDEX DATA IS 6
digits.target[67]
```

```python
# NOW LETS SEE IF OUR MODEL CAN PREDICT IT RIGHTLY
model.predict([digits.data[67]])
```

```python
model.score(X_test,Y_test)
```

```python
# NOW THAT WE KNOW THE SCORE OF THE MODEL TO FIND THE DATAS FOR WHICH OUR MODEL FAILED TO PREDICT CORRECTLY WE
# USE CONFUSION MATRIX 
# NOW LETS SEE THE PREDICTIONS MADE BY THE MODEL AND STORE IT IN Y_PREDICTED
Y_predicted=model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test, Y_predicted)
cm
```

```python
# NOW THAT WE KNOW THE SCORE OF THE MODEL TO FIND THE DATAS FOR WHICH OUR MODEL FAILED TO PREDICT CORRECTLY WE
# USE CONFUSION MATRIX 
# NOW LETS SEE THE PREDICTIONS MADE BY THE MODEL AND STORE IT IN Y_PREDICTED
Y_predicted=model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test, Y_predicted)
cm
import seaborn as sn
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')
# SO HERE 0 REPRESENTS THE VALUE PREDICTED WAS RIGHT AND MATCHES WITH ACTUAL VALUE 1 MEANS IT THE PREDICTED VALUE PREDICTED WRONGLY 1 TIME
# FOR THE GIVEN VALUE 
# THE DIAGONAL IS CONSISTS OF THE CORRECTLY REPRESENTED VALUES AND THE NUMBER OF TIMES IT WAS PREDICTED RIGHT
```
